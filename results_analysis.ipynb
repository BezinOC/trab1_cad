C Study Case

```python
df_row_column_c = pd.read_csv(r'results\multiplication_results_row_column_c.txt', sep=" ")
df_column_row_c = pd.read_csv(r'results\multiplication_results_column_row_c.txt', sep=" ")
```

```python
df_c = df_row_column_c.merge(df_column_row_c, on="size", suffixes=("_rc", "_cr")).set_index('size')
```

```python
# Plot C results

# Configure the plot style
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(20, 10))
plt.title('ComparaÃ§Ã£o de Resultados em C')
plt.xlabel('Ordem da Matrix (n)')
plt.ylabel('Tempo (s)')

# Plot the curves
plt.plot(df_c.index, df_c['result_rc'], label='Curve 1', linewidth=2)
plt.plot(df_c.index, df_c['result_cr'], label='Curve 2', linestyle='--', linewidth=2)

# Add legend, grid, and show the plot
plt.legend(['Iterando cada elemento Linha a Linha', 'Iterando cada elemento Coluna a Coluna'], 
           loc='upper left', fontsize='xx-large', frameon=True, facecolor='white', edgecolor='black')
plt.grid(True)
plt.show()
```
